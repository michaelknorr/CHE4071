```python
from scipy.integrate import solve_ivp
import numpy as np
from plotly.subplots import make_subplots
```

```python
def dhdt(t, h):
  qin=0.15
  return (qin-0.2*np.sqrt(2*9.8*h))
```

```python
h0=(0.1/0.2)**2/(2*9.8)
```

```python
sol=solve_ivp(dhdt, (0, 30), [h0], method='Radau')
```

```python
hplot=sol.y.tolist()
tplot=np.linspace(0, 30, 100)
```

```python
plot=make_subplots(rows=1, cols=1)
plot.add_scatter(x=tplot, y=hplot[0], mode='markers')
plot.update_layout(width=800, height=600, template='plotly_dark')
```

```python
Ac=1 #m2
Ah=0.2#m2
qin_ss=0.1 #m3/s
g=9.8#m/s2
h0=(qin_ss/Ah)**2/(2*g)#m
hsp=1.2*h0

Kc=29 #CONTROLLER GAIN
hsp
```

```python
def dhdt_series(t, h):
  h1, h2, h3=h
  qin=qin_ss+Kc*(hsp-h3)
  qout1=Ah*np.sqrt(2*g*h1)
  qout2=Ah*np.sqrt(2*g*h2)
  qout3=Ah*np.sqrt(2*g*h3)
  return [qin-qout1, qout1-qout2, qout2-qout3] #dh1, dh2, dh3
```

```python
sol2=solve_ivp(dhdt_series, (0, 100), [h0, h0, h0], method='Radau', dense_output=True)
tplot2=np.linspace(0, 100, 100)

h=sol2.y
```

```python
h_1=h[0]
h_2=h[1]
h_3=h[2]
```

```python
plot2=make_subplots(rows=1, cols=1)
plot2.add_scatter(x=tplot2, y=h_1, mode='lines')
plot2.add_scatter(x=tplot2, y=h_2, mode='lines')
plot2.add_scatter(x=tplot2, y=h_3, mode='lines')
plot2.update_layout(width=800, height=600, template='plotly_dark')
```
